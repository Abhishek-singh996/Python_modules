```python
x = np.random.normal(size=(2,3))
```

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(np.random.normal(size=1000),kind='kde')
plt.show()
```

## Binomial distributons

```python
x = np.random.binomial(n=10,p=0.5,size=10)
x
```

```python
sns.displot(np.random.binomial(n=10,p=0.5,size=10),kind='kde')
plt.show()
```

## Difference between normal and binomial distribution
### main difference is that normal distribution is continuous whereas binomial is discrete, but if there are enough data points it will be quite similar to normal distribution with certain loc and scale.

```python
dist = [np.random.normal(loc=50,scale=5,size=1000),np.random.binomial(n=100,p=0.5,size=1000)]
```

```python
sns.kdeplot(dist)
plt.legend(title='distributions',loc = 'upper right',labels=['normal','binomial'])
# sns.kdeplot()
plt.show()
```

## Poisson distributions

```python
sns.displot(np.random.poisson(lam=2,size=1000))
```

```python
dist = [np.random.normal(loc=50,scale=7,size=1000),np.random.poisson(lam=50,size=1000)]

sns.kdeplot(dist)
plt.legend(title='distributions',loc = 'upper right',labels=['normal','poisson'])
plt.show()
```

```python
## difference between binomial and poisson distribution
sns.kdeplot(np.random.binomial(n=1000,p=0.01,size=1000))
sns.kdeplot(np.random.poisson(lam=10,size=1000))
plt.legend(title='distributions',loc = 'upper right',labels=['binomial','poisson'])
plt.show()
```

## uniform distributions

```python
x = np.random.uniform(size=1000)
sns.displot(x,kind='kde')
plt.legend(labels=['uniform'])
plt.show()
```

## logistic Distributions

```python
x = random.logistic(loc=1,scale=2,size=(2,3))
x
```

```python
sns.displot(random.logistic(size=1000),kind='kde')
plt.legend(labels=['logistic distributions'])
plt.show()
```

```python
## Logistic v/s normal distributions
sns.kdeplot(random.logistic(size=1000))
sns.kdeplot(random.normal(scale=2,size=1000))
plt.legend(labels=['logistic','normal'])
plt.show()
```

## Multinomial distrinutions

```python
x = random.multinomial(n=6,pvals=[1/6,1/6,1/6,1/6,1/6,1/6])
x
```

## Exponential distribution

```python
x = random.exponential(scale=2,size=(2,3))
x
```

```python
sns.kdeplot(random.exponential(size=1000))
plt.legend(labels=['exponential distribution'])
plt.show()
```

## chi square distribution

```python
x = random.chisquare(df=2,size=(2,3))
x
```

```python
sns.kdeplot(random.chisquare(df=1,size=1000))
plt.legend(labels=['chi square distribution'])
plt.show()
```

## Rayleigh distribution

```python
x = random.rayleigh(scale=2,size=(2,3))
x
```

```python
sns.kdeplot(random.rayleigh(size=1000))
plt.legend(labels=['rayleigh distribution'])
plt.show()
```

```python
## similarity between rayleigh and chisquare distribution
sns.kdeplot(random.chisquare(df=2,size=1000))
sns.kdeplot(random.rayleigh(scale=1,size=1000))
plt.legend(labels=['chi square','rayleigh distribution'])
plt.show()
```

## Pareto distribution

```python
x = random.pareto(a=2,size=(2,3))
x
```

```python
sns.displot(random.pareto(a=2,size=1000))
plt.legend(labels=['pareto distribution'])
plt.show()
```

## zipf Distribution
